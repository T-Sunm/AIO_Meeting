```python
# Transpose

A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print("Before Transpose: \n", A)
print("After Transpose: \n", A.T)
```

```python
# ƒê·ªãnh th·ª©c c·ªßa ma tr·∫≠n 2√ó2
# det(A)=(1‚ãÖ4)‚àí(2‚ãÖ3)=4‚àí6=‚àí2

# ƒê·ªãnh th·ª©c ma tr·∫≠n 3x3
# det(B)=aei+bfg+cdh‚àíceg‚àíbdi‚àíafh

A = np.array([
    [1, 2],
    [3, 4],
])

print("Matrix", A)
print("determinant of A: \n", np.linalg.det(A))
```

```python
# ƒê·ªãnh nghƒ©a v√† √Ω nghƒ©a c·ªßa ma tr·∫≠n ngh·ªãch ƒë·∫£o (Inverse Matrix)
# det(A) = 0 => kh c√≥ ma tr·∫≠n ngh·ªãch ƒë·∫£o ,  
A = np.array([
    [1, 2],
    [4, 6]
])

print(A)
print("Determinant A", np.linalg.det(A))
B = np.linalg.inv(A)
print("Inverse A \n", B)


# ƒê·ªãnh nghƒ©a AB = BA = I
print("Dot Product A.B: \n", A.dot(B))
print("Dot Product B.A: \n", B.dot(A))
```

```python
# Cosine Similarity : Measure the similarity between two vectors

A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

print("A :", A)
print("B :", B)

# compute cosine similarity
cosine = A.dot(B) / (np.linalg.norm(A) * np.linalg.norm(B))
print("Cosine Similarity", cosine)
```

```python
from PIL import Image

image1 = Image.open('./data/image1.jpg')
image2 = Image.open('./data/image2.jpg')
query_image = Image.open('./data/query_image.jpg')

print(image1.size , image2.size , query_image.size) 
```

```python
image1 = np.array(image1) 
image2 = np.array(image2)
query_image = np.array(query_image)
```

```python
import matplotlib.pyplot as plt

image1_gray = image1.mean(axis=2) 
image2_gray = image2.mean(axis=2)
query_image_gray = query_image.mean(axis=2)
print(image1_gray)

plt.imshow(image1_gray, cmap="gray")
```

```python
flatten_image1 = image1_gray.flatten()
flatten_image2 = image2_gray.flatten()
flatten_query_image = query_image_gray.flatten()
print(flatten_query_image)
```

```python
Similar_img1_qrimage = flatten_image1.dot(
    flatten_query_image) / (np.linalg.norm(flatten_image1) * np.linalg.norm(flatten_query_image))
print("Similarity between image1 and query_image: ",Similar_img1_qrimage)

Similar_img2_qrimage = flatten_image2.dot(
    flatten_query_image) / (np.linalg.norm(flatten_image2) * np.linalg.norm(flatten_query_image))
print("Similarity between image2 and query_image: ", Similar_img2_qrimage)
```
