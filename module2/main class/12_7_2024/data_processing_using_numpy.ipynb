```python
arr = np.array([1, 2, 3, 4, 5])
print("arr_1c.shape: ", arr.shape)


# 1D -> 2D : thêm np.newaxis(thêm số 1) ở cuối mảng ->(5,) -> (5,1) 
arr_2c =arr[:,np.newaxis]
print(arr_2c)
print("arr_2c.shape: ", arr_2c.shape)


# 1D -> 2D
# 1D -> 2D : thêm np.newaxis(thêm số 1) ở đầu mảng ->(5,) -> (1, 5)
arr_2r = arr[np.newaxis,:]
print(arr_2r)
print("arr_2r.shape: ", arr_2r.shape)

# 2D -> 3D: thêm np.newaxis(thêm số 1) ở đầu mảng ->(1, 5) -> (1, 1, 5)
arr_3d = arr_2r[np.newaxis, :]
print(arr_3d)
print("arr_2c.shape: ", arr_3d.shape)
```

```python
#multiplication

x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])

print('data x:', x)
print('data y:', y)

# tích các phần tử tương ứng giữa x và y
print('method 1: \n',x*y)
print('method 2: \n', np.multiply(x,y))
```

```python
# division 
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])

#  phép chia các phần tử tương ứng x cho y
print('method 1: \n', x / y)
print('method 2: \n', x // y)
print('method 3: \n', np.divide(x, y))
```

```python
# sigmoid

def sigmoid(x):
    sm = 1 + np.exp(x * -1)

    return 1 / sm

data = np.array([1, 5, -4, 3, -2])

print(sigmoid(data))
```

```python
# Relu function

def relu(x):
    x = np.where(x < 0, 0, x)
    return x


data = np.array([1, 5, -4, 3, -2])

print(relu(data))
```

```python
import numpy as np
from urllib import request
import gzip
import pickle

filename = [
    ["training_images", "train-images-idx3-ubyte.gz"],
    ["test_images", "t10k-images-idx3-ubyte.gz"],
    ["training_labels", "train-labels-idx1-ubyte.gz"],
    ["test_labels", "t10k-labels-idx1-ubyte.gz"]
]

def download_mnist():
    folder = 'data_mnist/'
    base_url = "http://yann.lecun.com/exdb/mnist/"
    for data in filename:
        print("Downloading " + data[1] + "...")

        # urlretrieve(đường dẫn muốn lấy data, đường dẫn muốn lưu data ở máy)
        request.urlretrieve(base_url + data[1],folder + data[0])
        print("Download complete.")


download_mnist()
```

```python
data = np.array([[2, 6, 0, 1],
                 [3, 0, 8, 3],
                 [4, 4, 5, 0],
                 [2, 0, 4, 6]
                 ])

print("data \n", data)

# Hàm apply_along_axis trong NumPy được sử dụng để áp dụng một hàm cụ thể dọc theo một trục 
# (axis) của một mảng NumPy. Điều này rất hữu ích khi bạn muốn áp dụng một hàm tùy chỉnh lên các cột hoặc hàng của một mảng mà không cần phải viết vòng lặp thủ công.
result = np.apply_along_axis(np.max, axis=1, arr=data)
```

```python
img = cv2.imread('data/samples/cat.0.jpg', 1)

def compute_average(vector):
    result = vector.mean()
    result = result.astype(np.uint8)
    return result


gray_img = np.apply_along_axis(compute_average, axis=2, arr=img)

plt.imshow(gray_img, cmap="gray")
```
