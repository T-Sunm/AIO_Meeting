## Illustration

```python
import tensorflow as tf

opt = tf.keras.optimizers.SGD(learning_rate=0.1)
print(opt)

def compute_loss(x):
    return x**2
```

```python
list_grad = []
list_x = []

var = tf.Variable(70.0)
list_x.append(var.numpy())
#print('var: ', var.numpy())

lr = 0.1
for i in range(100):
    with tf.GradientTape() as tape:
        loss = compute_loss(var)

    grads = tape.gradient(loss, var)
    list_grad.append(grads.numpy())
    #print(i, '-grads: ', grads.numpy())

    var.assign_add(-lr*grads)
    list_x.append(var.numpy())
    #print('var: ', var.numpy())
    #print('\n')
```

```python
import matplotlib.pyplot as plt
%matplotlib qt

plt.plot(list_grad[80:], label='grads')
plt.plot(list_x[80:100], label='x')
plt.xlabel('iteration')
plt.ylabel('Value')
plt.legend()
```

## 2D Square Function

```python
import tensorflow as tf

opt = tf.keras.optimizers.SGD(learning_rate=0.1)
print(opt)

def compute_loss(var):
    return tf.reduce_sum(var*var)
```

```python
var = tf.Variable([6.0, 9.0])
print('var: ', var.numpy())

lr = 0.1
for i in range(4):
    with tf.GradientTape() as tape:
        loss = compute_loss(var)

    grads = tape.gradient(loss, var)
    print('grads: ', grads.numpy())

    var.assign_add(-lr*grads)
    print('var: ', var.numpy())
    print('\n')
```
