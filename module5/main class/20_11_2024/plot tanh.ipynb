```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -3 to 3
x = np.linspace(-0.5, 0.5, 100)

# Calculate tanh(x)
y = np.tanh(x)
y2 = x
# Create the plot
plt.plot(x, y)
plt.plot(x,y2)
# Add labels and title
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.title("Hyperbolic Tangent Function")

# Add grid
plt.grid(True)

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# Generate x values from -5 to 5
x = np.linspace(-0.5, 0.5, 100)

# Calculate sigmoid(x)
y = sigmoid(x)
y2 = 1/2 + x/4
# Create the plot
plt.plot(x, y)
plt.plot(x, y2)
# Add grid
plt.grid(True)

# Add labels and title
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Function")

# Display the plot
plt.show()
```
