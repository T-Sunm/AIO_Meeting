```python
import cv2
import numpy as np
import PIL.Image as Image
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
bone = cv2.imread("./bone.jpg")
```

```python
bone.shape
```

```python
bone = cv2.cvtColor(bone, cv2.COLOR_BGR2GRAY)
```

```python
bone.shape
```

```python
# Display original and filtered images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(bone, cmap='gray')
plt.axis("off")
```

```python
# Define Laplacian kernel
laplacian_kernel = np.array([[0,  1,  0],
                             [1, -4,  1],
                             [0,  1,  0]])
```

```python
# Apply the Laplacian filter
bone_laplacian = cv2.filter2D(bone, -1, laplacian_kernel)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# Sharpen the image by subtracting the Laplacian from the original image
sharpened_image = cv2.subtract(bone, bone_laplacian)
```

```python
# Display original, Laplacian, and sharpened images
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened_image, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# Apply Sobel filters
sobel_x = cv2.Sobel(sharpened_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(sharpened_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# Convert back to uint8
sobel_magnitude = np.uint8(sobel_magnitude)
```

```python
# Display original, Laplacian, sharpened, and Sobel-filtered images
plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 4)
plt.title("Sobel Filtered Image")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis("off")

plt.show()
```

```python
blurred_image = cv2.medianBlur(sobel_magnitude, 3)
```

```python
# Display original, Laplacian, sharpened, Sobel-filtered, and blurred images
plt.figure(figsize=(25, 5))
plt.subplot(1, 5, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 5, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.subplot(1, 5, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 5, 4)
plt.title("Sobel Filtered Image")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis("off")

plt.subplot(1, 5, 5)
plt.title("Blurred Image")
plt.imshow(blurred_image, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# Multiply sharpened and blurred images
combined_image = cv2.multiply(sharpened_image, blurred_image)
```

```python
# Display original, Laplacian, sharpened, Sobel-filtered, blurred, and combined images
plt.figure(figsize=(30, 5))
plt.subplot(1, 6, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 6, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.subplot(1, 6, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 6, 4)
plt.title("Sobel Filtered Image")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis("off")

plt.subplot(1, 6, 5)
plt.title("Blurred Image")
plt.imshow(blurred_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 6, 6)
plt.title("Combined Image")
plt.imshow(combined_image, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# Add the original image and the combined image
final_image = cv2.add(bone, combined_image)
```

```python
# Display original, Laplacian, sharpened, Sobel-filtered, blurred, combined, and final images
plt.figure(figsize=(35, 5))
plt.subplot(1, 7, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 7, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.subplot(1, 7, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 7, 4)
plt.title("Sobel Filtered Image")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis("off")

plt.subplot(1, 7, 5)
plt.title("Blurred Image")
plt.imshow(blurred_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 7, 6)
plt.title("Combined Image")
plt.imshow(combined_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 7, 7)
plt.title("Final Image")
plt.imshow(final_image, cmap='gray')
plt.axis("off")

plt.show()
```

```python
# Apply exponential transformation
c = 0.5  # Scaling factor
exp_transformed_image = np.uint8(c * (np.exp(final_image / 255.0) - 1) * 255.0 / (np.exp(1) - 1))
```

```python
# Display original, Laplacian, sharpened, Sobel-filtered, blurred, combined, final, and exp-transformed images
plt.figure(figsize=(40, 5))
plt.subplot(1, 8, 1)
plt.title("Original Image")
plt.imshow(bone, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 2)
plt.title("Laplacian Filtered Image")
plt.imshow(bone_laplacian, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 4)
plt.title("Sobel Filtered Image")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 5)
plt.title("Blurred Image")
plt.imshow(blurred_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 6)
plt.title("Combined Image")
plt.imshow(combined_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 7)
plt.title("Final Image")
plt.imshow(final_image, cmap='gray')
plt.axis("off")

plt.subplot(1, 8, 8)
plt.title("Exp Transformed Image")
plt.imshow(exp_transformed_image, cmap='gray')
plt.axis("off")

plt.show()
```
